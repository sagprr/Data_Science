```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://en.wikipedia.org/wiki/Randeep_Hooda')
soup = BeautifulSoup(res.text,'html.parser')
```

```python
heading = soup.find('h1').text
```

```python
heading
```

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'

corpus = corpus.strip()
corpus_f = corpus.format()
```

```python
print(corpus_f)
```
