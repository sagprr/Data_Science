```python
cols = ['DiabetesPedigreeFunction', 'BMI', 'Age', 'Pregnancies', 'Insulin', 'Glucose', 'BloodPressure']

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lbound = Q1 - 1.5 * IQR
    ubound = Q3 + 1.5 * IQR

    # Apply the mask to filter out outliers, or however you intend to use it
    mask = (df[col] >= lbound) & (df[col] <= ubound)


```

```python
X = df[mask]
Y = df[mask]
```
